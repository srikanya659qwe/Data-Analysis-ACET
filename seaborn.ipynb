### How to read Datasets in seaborn:
        - sns.load_dataset('dataset name')

```python
iris=sns.load_dataset('iris')
iris
```

### catplot():
    - by defualt it returns scatter plot
            - sns.catplot()

```python
sns.catplot(x='species',y='sepal_length',data=iris,palette='winter')
```

```python
sns.catplot(x='species',y='sepal_length',data=iris,palette='winter',kind='box')
```

```python
sns.catplot(x='species',y='sepal_length',data=iris,palette='winter',orient='h')
```

```python
sns.catplot(data=iris,palette='winter',orient='h',kind='box')
```

### joint plot():
        - it is a combination two plots(scatter,histogram)
        - sns.jointplot()

```python
sns.jointplot(x='petal_width',y='sepal_width',data=iris)
```

```python
sns.jointplot(x='petal_width',y='sepal_width',data=iris,color='g')
```

```python
sns.jointplot(x='petal_width',y='sepal_width',data=iris,kind='hex')
```

#### pairplot():
        - it displays multiple plots at a time in single graph
        - by default it returns scatter plot
        - sns.pairplot()

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species')
```

```python
sns.pairplot(iris,hue='sepal_length',palette="spring")
```
