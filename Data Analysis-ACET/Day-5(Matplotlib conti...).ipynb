#### linear/line plot:

```python
x_axis=[10,12,14,16]
y_axis=[5,6,7,8]
plt.plot(x_axis,y_axis)
```

```python
x_axis=[10,12,14,16]
y_axis=[5,6,7,8]
plt.plot(x_axis,y_axis,label="Graph")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Linear Graph")
plt.legend(loc="upper left")
```

```python
import numpy as np
```

```python
a=np.linspace(0,5,20)
a
```

```python
plt.plot(a,a**2,color='red',label="Red")
plt.plot(a,a**3,color='green',label="Green")
plt.plot(a,a**4,color='blue',label="Blue")
plt.legend(loc="upper left")
plt.show()
```

#### linestyles:

```python
x1=np.arange(20)
x1
```

```python
plt.plot(x1,x1+10,linestyle='--',color='red')
plt.plot(x1,x1+15,linestyle=':',color='green')
plt.plot(x1,x1+20,linestyle='-.',color='yellow')
plt.plot(x1,x1+25,linestyle='-',color='pink')
```

#### Marker:
        -o,<,>,^,D,H


```python
x2=np.linspace(10,30,50)
x2
```

```python
plt.plot(x2,np.sin(x2),marker='o',color='green',markersize=8,markerfacecolor="yellow")
plt.plot(x2,np.cos(x2),marker='^',color='red',markersize=8,markerfacecolor="pink")
```

```python
plt.plot(x2,np.sin(x2-1))
plt.plot(x2,np.sin(x2-2))
plt.plot(x2,np.sin(x2-3))
plt.plot(x2,np.sin(x2-4))
plt.plot(x2,np.sin(x2-5))
plt.plot(x2,np.sin(x2-6))
plt.plot(x2,np.sin(x2-7))
```

#### BarPlot():
    - plt.bar()

```python
Year=['2023 Dec','2022 Dec','2021 Dec','2020 Dec']
Gold_price=[70000,65000,60000,55000]
plt.bar(Year,Gold_price,color='red')
```

```python
Year=['2023 Dec','2022 Dec','2021 Dec','2020 Dec']
Gold_price=[70000,65000,60000,55000]
Silver_price=[25000,30000,35000,40000]
plt.bar(Year,Gold_price,color='red',width=0.33)
plt.bar(Year,Silver_price,color='green',width=0.33)
```

#### Boxplot():
    - it gives a summary of numerical data through quartiles
            - min
            - quartiles
                    - lower quartile
                    - mediun
                    - upper quartile
            - max
            
    - syntax:
           plt.boxplot()

```python
numbers=np.random.randint(0,10000,1000)
numbers
```

```python
plt.boxplot(numbers)
```

```python
plt.boxplot(numbers,notch=True)
```

```python
product_sale1=np.random.randint(0,5000,50)
product_sale2=np.random.randint(0,7000,100)
product_sale3=np.random.randint(1000,3500,500)
plt.boxplot([product_sale1,product_sale2,product_sale3],labels=['first','second','third'],patch_artist=True,boxprops=dict(facecolor="pink",color="red"))
```
